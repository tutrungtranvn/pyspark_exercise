```python
import pyspark
from pyspark.sql.types import *
```

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('Pyspark_test').master('local[*]').getOrCreate()
```

```python
df = spark.read.csv('netflix_titles.csv')
```

```python
df = spark.read.option('header','true').csv('netflix_titles.csv')
```

```python
df = df.drop('description')
```

```python
df.show()
```

```python
df.toPandas()
```

```python
#check Schema
df.printSchema()
```

```python
df.describe()
```

```python
from pyspark.sql.functions import count, when, isnan, col
```

```python
df.columns
```

```python
df.select([count(when((col(c) == '') |isnan(c) | col(c).isNull(), c)).alias(c) for c in df.columns]).show()
```

```python
df.filter("duration is Null").toPandas()
```

```python
df = df.where(df['title'].isNotNull()).where(df['listed_in'].isNotNull())
```

```python
df.toPandas()
```

```python
df.select('director').show()
```

```python
#Add columns 
df.withColumn('future_year',df['release_year']+2).toPandas()
```
